```python
# test
steps = 5
eps = 5
maxDepth = 5
```

```python
time = 0
solved = 0
for i in range(eps):
    
    moves = generateScramble(steps) # 4 move scramble
    c = cube()
    c.move(moves)
    print(moves)
    c.view()
    
    s = solver(c, h=compound1)
    start = timeit.default_timer()
    s.solve(maxDepth=maxDepth)
    if s.solutionFound:
        solved +=1
    stop = timeit.default_timer()
    timeTaken = stop - start
    time += timeTaken
    print('Time: ', timeTaken)
    print('=================================\n')
    
print('Average time taken: {}'.format(time/eps))
print('Solved: {}/{}'.format(solved, eps))
```
