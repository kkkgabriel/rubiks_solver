```python
from rubiks import *
from solver import *
import timeit
```

```python
steps = 4
moves = generateScramble(steps) # 4 move scramble
c = cube()
c.move(moves)
print(moves)
c.view()
```

```python
# counting the number of correct colors on face
def numberOfCorrectColorsOnFace(queueItem):
    cube = queueItem[0]
    score = 0
    for face in cube.faces:
        center = face[4]
        score += list(face.values()).count(center)
    return score
```

```python
s = solver(c, h=numberOfCorrectColorsOnFace)
start = timeit.default_timer()
s.solve(maxDepth=steps)
stop = timeit.default_timer()
print('Time: ', stop - start)
```

```python
# brute force
s = solver(c)
start = timeit.default_timer()
s.solve(maxDepth=steps)
stop = timeit.default_timer()
print('Time: ', stop - start)  
```
