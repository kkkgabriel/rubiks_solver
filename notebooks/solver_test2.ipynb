```python
from rubiks import *
from solver import *
import timeit
```

```python
# counting the number of correct colors on face
def numberOfCorrectColorsOnFace(queueItem):
    cube = queueItem[0]
    score = 0
    for face in cube.faces:
        center = face[4]
        score += list(face.values()).count(center)
    return score
```

```python
steps = 5
eps = 5
maxDepth = 5
```

```python
time = 0
for i in range(eps):
    
    moves = generateScramble(steps) # 4 move scramble
    c = cube()
    c.move(moves)
    print(moves)
    c.view()
    
    s = solver(c, h=numberOfCorrectColorsOnFace)
    start = timeit.default_timer()
    s.solve(maxDepth=maxDepth)
    stop = timeit.default_timer()
    timeTaken = stop - start
    time += timeTaken
    print('Time: ', timeTaken)
    print('=================================\n')
    
print('Average time taken: {}'.format(timeTaken/eps))
```
