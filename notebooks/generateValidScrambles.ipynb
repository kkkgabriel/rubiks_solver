```python
from rubiks import *
from solver import *
from heuristics import *
import timeit
import pandas as pd
```

```python
h = {
    "numberOfCorrectColorsOnFace": numberOfCorrectColorsOnFace,
    "numberOfPairedCornerNEdges": numberOfPairedCornerNEdges,
    "oneStepCornerEdgePair": oneStepCornerEdgePair,
    "compound1": compound1,
    "compound2": compound2
}
```

```python
all_steps = [3,5,7,10]
# all_steps = [3,5]
eps = 20
```

```python
scrambles = {steps: {} for steps in all_steps}
# make scrambles
for steps in all_steps:
    for i in range(eps):
        found = False
        while not found:
            scramble = generateScramble(steps)
            c = cube()
            c.move(scramble)
            s = solver(c, h=numberOfPairedCornerNEdges)
            s.solve(maxDepth=steps, display=False, verbose=False, earlyTermination=500)
            if s.solutionFound and len(s.solution) == steps:
                scrambles[steps][scramble] = s.solution
                found = True
    
```
