# Corner-edge pairing

```python
top_pairs = [
    [[6,7], [18,19]],
    [[7,8], [19,20]],
    [[5,6],[9,10]],
    [[2,5],[10,11]],
    [[1,2],[45,46]],
    [[0,1],[46,47]],
    [[0,3],[36,37]],
    [[3,6],[37,38]]
]
mid_pairs = [
    [[9,12],[20,24]],
    [[23,26],[12,15]],
    [[45,48],[11,14]],
    [[48,51],[14,17]],
    [[47,50],[36,39]],
    [[50,53],[39,42]],
    [[38,41],[18,21]],
    [[41,44],[21,24]]
]
bottom_pairs = [
    [[27,28],[24,25]],
    [[28,29],[25,26]],
    [[29,32],[15,16]],
    [[32,35],[16,17]],
    [[34,35],[51,52]],
    [[33,34],[52,53]],
    [[30,33],[42,43]],
    [[27,30],[43,44]]
]
```

### top pairs

```python
s = list('n' * 54)

p = []
for pairs in top_pairs:
    for pair in pairs:
        p += pair

for i in p:
    s[i] = 'w'


s = ''.join(s)
c = cube(colours=s)
c.view()
```

# center edge pairing

```python
top_middle_centers = [ # uf, ul, ub, ur
    [4,22], [4,40], [4,49],[4,13]
]
middle_centers = [ # fl,lb,br,rf
    [13,22], [22,40], [40,49], [49,13],
]
bottom_middle_centers = [ # df, dl, db, dr
    [31,22], [31,40], [31,49],[31,13]
]
top_edges = [
    [7,19],[3,37],[1,46],[5,10]
]
side_edges = [
    [12,23],[21,41],[39,50],[48,14] # clockwise from top
]
bottom_edges = [
    [28,25],[30,43],[34,52], [32,16]
]
```
