```python
class solver():
    def __init__(self, startCube, h = lambda x: return 1):
        self.startCube = startCube
        self.h = h
        self.queue = [(startCube.duplicate(), 0)]
        self.solution = ''
        self.solutionFound = False
        
    def solve(self, maxDepth=6):
        while self.solutionFound:
            queueItem = self.pop()
            cube = queueItem[0]
            depth = queueItem[1]
            if cube.checkSolved():
                self.solution = cube.getMoves()
                self.solutionFound = True
            else:
                for m in MOVES:
                    self.insert((cube.duplicate().move(m), depth+1)
        
    def insert(self, cube):
        self.queue.append(cube)
        self.queue.sort(reverse=True, key=self.h)
        
    def pop(self):
        cube = self.queue[0]
        self.queue = self.queue[1:]
        return cube
```

```python
def h(queueItem):
    cube = queueItem[0]
    score = 0
    for face in cube.faces:
        center = face[4]
        score += list(face.values()).count(center)
    return score
```
