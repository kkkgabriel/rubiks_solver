```python
top_middle_centers = [ # uf, ul, ub, ur
    [4,22], [4,40], [4,49],[4,13]
]
middle_centers = [ # fl,lb,br,rf
    [13,22], [22,40], [40,49], [49,13],
]
bottom_middle_centers = [ # df, dl, db, dr
    [31,22], [31,40], [31,49],[31,13]
]
top_edges = [
    [7,19],[3,37],[1,46],[5,10]
]
side_edges = [
    [12,23],[21,41],[39,50],[48,14] # clockwise from top
]
bottom_edges = [
    [28,25],[30,43],[34,52], [32,16]
]
```

```python
# defining middleEdgePairing
def middleEdgePairing(queueItem):
    cube = queueItem[0]
    score = 0
    state = cube.getState()
    centersAndEdges = [
        [top_middle_centers, top_edges],
        [middle_centers, side_edges],
        [bottom_middle_centers, bottom_edges]
    ]
    for centerAndEdge in centersAndEdges:
        (centers, edges) = centerAndEdge
        for center, edge in zip(centers, edges):
            if state[center[0]] == state[edge[0]] and state[center[1]] == state[edge[1]]:
                score +=1
    return score
```

```python
c = cube()
# c.R()
c.view()
middleEdgePairing((c, None))
```

## Test

```python
steps = 5
eps = 5
maxDepth = 5
```

```python
time = 0
solved = 0
for i in range(eps):
    
    moves = generateScramble(steps) # 4 move scramble
    c = cube()
    c.move(moves)
    print(moves)
    c.view()
    
    s = solver(c, h=middleEdgePairing)
    start = timeit.default_timer()
    s.solve(maxDepth=maxDepth)
    if s.solutionFound:
        solved +=1
    stop = timeit.default_timer()
    timeTaken = stop - start
    time += timeTaken
    print('Time: ', timeTaken)
    print('=================================\n')
    
print('Average time taken: {}'.format(time/eps))
print('Solved: {}/{}'.format(solved, eps))
```
