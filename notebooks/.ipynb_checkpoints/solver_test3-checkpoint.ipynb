```python
from rubiks import *
from solver import *
from heuristics import *
import timeit
```

```python
# counting number of top corner-edge pairs matched
top_pairs = [
    [[6,7], [18,19]],
    [[7,8], [19,20]],
    [[5,6],[9,10]],
    [[2,5],[10,11]],
    [[1,2],[45,46]],
    [[0,1],[46,47]],
    [[0,3],[36,37]],
    [[3,6],[37,38]]
]
mid_pairs = [
    [[9,12],[20,24]],
    [[23,26],[12,15]],
    [[45,48],[11,14]],
    [[48,51],[14,17]],
    [[47,50],[36,39]],
    [[50,53],[39,42]],
    [[38,41],[18,21]],
    [[41,44],[21,24]]
]
bottom_pairs = [
    [[27,28],[24,25]],
    [[28,29],[25,26]],
    [[29,32],[15,16]],
    [[32,35],[16,17]],
    [[34,35],[51,52]],
    [[33,34],[52,53]],
    [[30,33],[42,43]],
    [[27,30],[43,44]]
]
corner_edge_pairs = top_pairs + mid_pairs + bottom_pairs
def numberOfPairedCornerNEdges(queueItem):
    cube = queueItem[0]
    score = 0
    state = cube.getState()
    for pair in corner_edge_pairs:
        (p1, p2) = pair
        (n1, n2) = p1
        (n3, n4) = p2
        if state[n1] == state[n2] and state[n3] == state[n4]:
            score += 1
    return score
```

```python
steps = 5
eps = 5
maxDepth = 5
```

```python
time = 0
for i in range(eps):
    
    moves = generateScramble(steps) # 4 move scramble
    c = cube()
    c.move(moves)
    print(moves)
    c.view()
    
    s = solver(c, h=numberOfPairedCornerNEdges)
    start = timeit.default_timer()
    s.solve(maxDepth=maxDepth)
    stop = timeit.default_timer()
    timeTaken = stop - start
    time += timeTaken
    print('Time: ', timeTaken)
    print('=================================\n')
    
print('Average time taken: {}'.format(timeTaken/eps))
```
