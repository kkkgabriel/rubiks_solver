```python
'''
# of solid rows 
(a row is solid of it contains at least 2 of the same color) - from all perspectives (vertical & horizontal)

Horizontal Rows: 012, 345, 678
Vertical Rows: 036, 147, 258
'''
def numberOfSolidRows1(queueItem):
    # Row mappings
    rows = [[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8]]
    cube = queueItem[0]
    score = 0
    for face in cube.faces:
        for row in rows:
            count = set()
            for tile in row:
                count.add(face[tile])
            if len(count) <=2: # If at least 2 of the same color
                score += 1
    return score
```

```python
'''
# of solid rows 
(a row is solid of it contains all of the same color) - from all perspectives (vertical & horizontal)

Horizontal Rows: 012, 345, 678
Vertical Rows: 036, 147, 258
'''

def numberOfSolidRows2(queueItem):
    # Row mappings
    rows = [[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8]]
    cube = queueItem[0]
    score = 0
    for face in cube.faces:
        for row in rows:
            count = set()
            for tile in row:
                count.add(face[tile])
            if len(count) == 1:
                score += 1
    return score
```

```python
'''
# of solid rows 
With different tier for scoring:
- Same color for every tile in a row: 10 pints
- 2 colors for tiles in a row: 5 points
- 3 colors for tiles in a row: - 5 points (penalty)
'''

def numberOfSolidRows3(queueItem):
    # Row mappings
    rows = [[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8]]
    cube = queueItem[0]
    score = 0
    for face in cube.faces:
        for row in rows:
            count = set()
            for tile in row:
                count.add(face[tile])
            if len(count) ==1:
                score += 10
            if len(count) ==2:
                score += 5
            if len(count) ==3:
                score -= 5
    return score
```

```python
'''
# of solid rows 
With different tier for scoring:
- Same color for every tile in a row: 10 pints
- 2 colors for tiles in a row: 5 points
- 3 colors for tiles in a row: - 5 points (penalty)

With added incentive and alignment for rows
(found out to be not of good use)
'''

def numberOfSolidRows4(queueItem):
    # Row mappings
    hor_rows = [[0,1,2],[3,4,5],[6,7,8]]
    vert_rows = [[0,3,6],[1,4,7],[2,5,8]]
    
    cube = queueItem[0]
    score = 0
    for face in cube.faces:
        alignment = 0
        for row in hor_rows:
            count = set()
            for tile in row:
                count.add(face[tile])
            if len(count) ==1:
                score += 10
                alignment += 1
            if len(count) ==2:
                score += 5
                alignment += 1
            if len(count) ==3:
                score -= 5
        for row in hor_rows:
            count = set()
            for tile in row:
                count.add(face[tile])
            if len(count) ==1:
                score += 10
                alignment += 1
            if len(count) ==2:
                score += 5
                alignment += 1
            if len(count) ==3:
                score -= 5
        if alignment >= 3:
            score += 5
                
        
        
    return score
```

```python
'''
# of solid rows 
With different tier for scoring:
- Same color for every tile in a row: 3 pints
- 2 colors for tiles in a row: 2 points
- 3 colors for tiles in a row: - 2 points (penalty)
'''

def numberOfSolidRows5(queueItem):
    # Row mappings
    rows = [[0,1,2],[3,4,5],[6,7,8],[0,3,6],[1,4,7],[2,5,8]]
#     rows = [[0,1,2],[3,4,5],[6,7,8]]
    cube = queueItem[0]
    score = 0
    for face in cube.faces:
        for row in rows:
            count = set()
            for tile in row:
                count.add(face[tile])
            if len(count) ==1: # All same color
                score += 3
            if len(count) ==2: # 2 colors
                score += 2
            if len(count) ==3: # 3 colors
                score += -2
    return score
```

```python
# test if heuristic can solve the cube
s = solver(c, h=numberOfSolidRows)
start = timeit.default_timer()
s.solve(maxDepth=steps)
stop = timeit.default_timer()
print('Time: ', stop - start)
```

```python
steps = 5
eps = 5
maxDepth = 5
```

```python
# run through more iterations of the first heuristic
time = 0
for i in range(eps):
    
    moves = generateScramble(steps) # 4 move scramble
    c = cube()
    c.move(moves)
    print(moves)
    c.view()
    
    s = solver(c, h=numberOfSolidRows3)
    start = timeit.default_timer()
    s.solve(maxDepth=maxDepth)
    stop = timeit.default_timer()
    timeTaken = stop - start
    time += timeTaken
    print('Time: ', timeTaken)
    print('=================================\n')
    
print('Average time taken: {}'.format(time/eps))
```
