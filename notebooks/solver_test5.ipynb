```python
from rubiks import *
from solver import *
from heuristics import *
import timeit
```

```python
def potentialEdgeCornerPairs(queueItem):
    return oneStepCornerEdgePair(queueItem) + numberOfPairedCornerNEdges(queueItem)
```

```python
c = cube()
c.R()
c.view()
potentialEdgeCornerPairs((c, None))
```

```python
steps = 5
eps = 5
maxDepth = 5
```

```python
time = 0
solved = 0
for i in range(eps):
    
    moves = generateScramble(steps) # 4 move scramble
    c = cube()
    c.move(moves)
    print(moves)
    c.view()
    
    s = solver(c, h=numberOfPairedCornerNEdges)
    start = timeit.default_timer()
    if s.solve(maxDepth=maxDepth):
        solved +=1
    stop = timeit.default_timer()
    timeTaken = stop - start
    time += timeTaken
    print('Time: ', timeTaken)
    print('=================================\n')
    
print('Average time taken: {}'.format(timeTaken/eps))
print('Solved: {}/{}'.format(solved, eps))
```
